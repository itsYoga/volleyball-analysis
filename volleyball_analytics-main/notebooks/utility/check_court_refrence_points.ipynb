```python
import cv2
import json
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [15, 10]
```

```python
js = json.load(open('conf/court.json'))
```

```python
vid = cv2.VideoCapture('data/raw/videos/train/11.mp4')
status, frame = vid.read()

for item in js['11.mp4']['court']:
    frame = cv2.circle(frame, tuple(item), 5, (255, 0, 255), -1)

for item in js['11.mp4']['attack']:
    frame = cv2.circle(frame, tuple(item), 5, (0, 0, 255), -1)

for item in js['11.mp4']['center']:
    frame = cv2.circle(frame, tuple(item), 5, (0, 255, 255), -1)


frame = cv2.cvtColor(frame, 4)

plt.imshow(frame)
```
